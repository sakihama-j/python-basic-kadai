```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"名前は{self.name}で年齢は{self.age}です。大人です。")
    else:
      print(f"{self.name}は大人ではないです。")

human_names = ["ジェイニー", "クレア", "ソフィア"]
human_ages = [23, 22, 19]

i = 0
human_list = []

for name in human_names:
  age = human_ages[i]
  new_human = Human(name, age)
  human_list.append(new_human)

  i = i+1

for human in human_list:
  human.check_adult()
```
